```python
from collections import Counter
import re

word_list = []

# 파일 열기 (r: 읽기 모드)
with open("Avengers.txt", "r") as f:
    # 파일 내용 읽기
    # content = f.read()
    # print(content)

    # 한 줄씩 읽기
    for line in f:
        line_str = line.strip().lower()     # 소문자로 변환
        # print(line_str)
        # 특수문자 제거
        pattern = r"[^a-zA-Z0-9 ']"     # 알파벳, 숫자, 공백, 어포스트로피 제외한 나머지 모든 문자
        line_str = re.sub(pattern, '', line_str)

        line_list = line_str.split()    # 공백을 기준으로 분리해서 리스트로 저장
        word_list += line_list      # word_list에 추가하기

# print(word_list)

# # 특정 단어가 얼마나 포함되어 있는지 확인하기 위한 테스트용 코드 블럭
# special_word = 'tesseract'
# special_word_list = []
# for word in word_list:
#     if special_word in word:
#         print(word)
#         special_word_list.append(word)

# print(len(special_word_list))
# print(special_word_list)

bigram_list = []
# 리스트 컴프리헨션을 사용해 튜플 형태의 bi-gram으로 묶은 리스트 생성
bigram_list = [(word_list[i], word_list[i+1]) for i in range(len(word_list)-1)]
# 컴프리헨션 사용하지 않고 다음과 같이 작성해도 됨
# for i in range(len(word_list)-1):
#     bigram_list.append((word_list[i], word_list[i+1]))

# # 특정 바이그램 튜플이 얼마나 포함되어 있는지 확인하기 위한 테스트용 코드 블럭
# compare_tuple = ('the', 'tesseract')
# compare_list = []
# for bigram in bigram_list:
#     if compare_tuple == bigram:
#         compare_list.append(bigram)

# print(len(compare_list))
# print(compare_list)

bigram_counts = Counter(bigram_list)
print(bigram_counts.most_common(5))     # 가장 빈도수 높은 순으로 5개 출력
```
